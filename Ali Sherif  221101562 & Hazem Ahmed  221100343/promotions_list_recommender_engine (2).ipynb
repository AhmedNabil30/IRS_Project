# **Feature Engineering for Users, Promotions, and Interactions**

```python

user_interactions = interactions_df.groupby('User_ID').agg(
    total_interactions=('Interaction_ID', 'count'),
    avg_transaction_amount=('Transaction_Amount', 'mean'),
    unique_promos=('Promo_ID', 'nunique')
).reset_index()


users_features = pd.merge(users_df, user_interactions, on='User_ID', how='left')


users_features['total_interactions'].fillna(0, inplace=True)
users_features['avg_transaction_amount'].fillna(0, inplace=True)
users_features['unique_promos'].fillna(0, inplace=True)


promo_interactions = interactions_df.groupby('Promo_ID').agg(
    total_redemptions=('Interaction_ID', 'count'),
    avg_transaction_amount=('Transaction_Amount', 'mean')
).reset_index()


promotions_features = pd.merge(promotions_df, promo_interactions, on='Promo_ID', how='left')
promotions_features['total_redemptions'].fillna(0, inplace=True)
promotions_features['avg_transaction_amount'].fillna(0, inplace=True)

rfm_features = interactions_df.groupby(['User_ID', 'Promo_ID']).agg(
    recency=('Timestamp', lambda x: (pd.Timestamp('2025-01-20') - pd.to_datetime(x).max()).days),
    frequency=('Interaction_ID', 'count'),
    monetary=('Transaction_Amount', 'sum')
).reset_index()

interaction_features = pd.merge(interactions_df, rfm_features, on=['User_ID', 'Promo_ID'], how='left')

print("User-Level Features:")
print(users_features.head(), "\n")

print("Promotion-Level Features:")
print(promotions_features.head(), "\n")

print("Interaction-Level Features (RFM):")
print(interaction_features.head(), "\n")
```

# **Content-Based Recommendation System**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


promotions_features = promotions_df.copy()
promotions_features['combined_features'] = promotions_features[
    ['Category', 'Offer_Type', 'Payment_Method_Eligibility', 'Store_Location']
].apply(lambda x: ' '.join(x), axis=1)


vectorizer = TfidfVectorizer()
promo_vectors = vectorizer.fit_transform(promotions_features['combined_features'])
promo_similarity = cosine_similarity(promo_vectors)

def recommend_promotions(user_preferences, top_n=5):

    user_vector = vectorizer.transform([user_preferences])
    similarity_scores = cosine_similarity(user_vector, promo_vectors).flatten()
    similar_promos = similarity_scores.argsort()[-top_n:][::-1]
    return promotions_df.iloc[similar_promos]

print("Content-Based Recommendations for 'Groceries':")
print(recommend_promotions("Groceries"), "\n")
```

# **Collaborative Filtering with SVD**

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy


reader = Reader(rating_scale=(1, 5))
interaction_data = Dataset.load_from_df(
    interactions_df[['User_ID', 'Promo_ID', 'Transaction_Amount']], reader
)

trainset, testset = train_test_split(interaction_data, test_size=0.2)

collab_model = SVD()
collab_model.fit(trainset)

user_id = 1
promo_id = 10
prediction = collab_model.predict(user_id, promo_id)
print(f"Predicted rating for User {user_id} and Promo {promo_id}: {prediction.est:.2f}")

predictions = collab_model.test(testset)
print("Collaborative Filtering RMSE:", accuracy.rmse(predictions))
```

# **Context-Aware Feature Engineering**

```python

interactions_df['Timestamp'] = pd.to_datetime(interactions_df['Timestamp'])

def map_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'


interactions_df['Season'] = interactions_df['Timestamp'].dt.month.apply(map_season)

context_features = pd.merge(interactions_df, users_df[['User_ID', 'Location']], on='User_ID')
context_features = pd.merge(context_features, promotions_df[['Promo_ID', 'Store_Location']], on='Promo_ID')

print("Context-Aware Features:")
print(context_features.head())
```

# **Neural Collaborative Filtering Model**

```python
from keras.models import Model
from keras.layers import Input, Embedding, Flatten, Dot, Dense


num_users = users_df['User_ID'].nunique()
num_promos = promotions_df['Promo_ID'].nunique()

user_input = Input(shape=(1,))
user_embedding = Embedding(num_users, 50)(user_input)
user_vec = Flatten()(user_embedding)

promo_input = Input(shape=(1,))
promo_embedding = Embedding(num_promos, 50)(promo_input)
promo_vec = Flatten()(promo_embedding)

dot_product = Dot(axes=1)([user_vec, promo_vec])
output = Dense(1, activation='sigmoid')(dot_product)

model = Model([user_input, promo_input], output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

# **Evaluation Metrics for Recommendation Systems**

```python
import numpy as np

def precision_at_k(recommended, actual, k=5):
    recommended_at_k = recommended[:k]
    hits = sum([1 for promo in recommended_at_k if promo in actual])
    return hits / k

def recall_at_k(recommended, actual, k=5):
    recommended_at_k = recommended[:k]
    hits = sum([1 for promo in recommended_at_k if promo in actual])
    return hits / len(actual) if len(actual) > 0 else 0


def ndcg_at_k(recommended, actual, k=5):
    recommended_at_k = recommended[:k]
    dcg = sum([1 / np.log2(idx + 2) if promo in actual else 0 for idx, promo in enumerate(recommended_at_k)])
    idcg = sum([1 / np.log2(idx + 2) for idx in range(min(len(actual), k))])
    return dcg / idcg if idcg > 0 else 0


actual_promotions = [5, 10, 20]
recommended_promotions = [5, 15, 10, 25, 30]

precision = precision_at_k(recommended_promotions, actual_promotions)
recall = recall_at_k(recommended_promotions, actual_promotions)
ndcg = ndcg_at_k(recommended_promotions, actual_promotions)

print(f"Precision@K: {precision:.2f}")
print(f"Recall@K: {recall:.2f}")
print(f"NDCG@K: {ndcg:.2f}")
```

# **Recommendation System GUI (using gradio)**

```python
import gradio as gr
import pandas as pd


users_df = pd.read_csv('walmart_users.csv')
promotions_df = pd.read_csv('walmart_promotions.csv')

def hybrid_recommendations(user_id, top_n=5):
    recommended_promos = promotions_df.sample(top_n)
    return recommended_promos[['Promo_ID', 'Category', 'Discount', 'Offer_Type']]

def recommend(user_id, top_n):
    recommendations = hybrid_recommendations(user_id, top_n)
    return recommendations

user_ids = list(users_df['User_ID'].unique())
demo = gr.Interface(
    fn=recommend,
    inputs=[
        gr.Dropdown(user_ids, label="Select User ID"),
        gr.Slider(1, 10, value=5, step=1, label="Number of Recommendations"),
    ],
    outputs=gr.Dataframe(headers=["Promo_ID", "Category", "Discount", "Offer_Type"]),
    title="Promotion List Recommender System",
)

demo.launch()
```
